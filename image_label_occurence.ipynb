```python
import matplotlib.pyplot as plt
import os
import matplotlib
```

```python
input_path = './result/travel/labels'
```

```python
label_counts = {}
```

```python
os.listdir(input_path)
```

```python
for user in os.listdir(input_path):
    print(user)
    for filename in os.listdir(os.path.join(input_path, user)):
        if filename.endswith(".txt"):
            file = open(os.path.join(input_path , user, filename), 'r')
            while True:
                label = file.readline()
                if not label:
                    break
                elif label.startswith('Labels:'):
                    continue
                label = label.rstrip()
                if label in label_counts:
                    label_counts[label] += 1
                else:
                    label_counts[label] = 1
```

```python
label_counts
```

```python
sorted_labels, sorted_counts = zip(*sorted(label_counts.items(), key=lambda x: x[1], reverse=True)[:20])
```

```python
matplotlib.rcParams['font.size'] = 15
matplotlib.rcParams['font.family'] = 'appleGothic'
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
plt.figure(figsize=(20, 10))
plt.bar(sorted_labels, sorted_counts)
plt.xlabel('라벨')
plt.ylabel('횟수')
plt.title('라벨별 빈도 #travel')
plt.xticks(rotation=45, ha="right")
plt.savefig('./result/barchart_travel.png', bbox_inches='tight')
plt.show()
```

```python
from wordcloud import WordCloud
```

```python
# Generate word cloud
wordcloud = WordCloud(width=1000, height=1000, background_color='white', max_words=100).generate_from_frequencies(label_counts)

# Display the generated image:
plt.figure(figsize=(10, 10))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.tight_layout(pad=0)
plt.savefig('./result/wordcloud_travel.png')
plt.show()
```
